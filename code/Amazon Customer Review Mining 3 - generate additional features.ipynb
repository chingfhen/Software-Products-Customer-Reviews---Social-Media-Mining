```python
import pandas as pd
import numpy as np
import string 
from collections import Counter
punctuations = string.punctuation 
```

```python
path = r"C:\Users\tanch\Documents\NTU\NTU Year 4\Semester 1\CS4022 - Social Media Mining\Assignments\Assignment 2 - Customer Review Mining Project\data\dataset.csv"
dataset = pd.read_csv(path)
dataset.head()
```

```python
def get_num_uppercase_words(text):
    count = 0
    for word in text:
        if word.isupper():
            count+=1
    return count

def get_num_punctuations(text):
    output = {}
    counts = Counter(text)
    for punc in punctuations:
        output[punc] = counts.get(punc,0)
    return output
```

### generate additional features
- number of words
- number of fully upper case words
- ratio of fully upper case words
- number of every punctuation

```python
additional_features = pd.DataFrame()
additional_features['num_words'] = dataset["reviewText"].apply(lambda text: len(text.split()))
additional_features['num_uppercase_words'] = dataset["reviewText"].map(get_num_uppercase_words)
punctuation_counts_df = pd.DataFrame(list(map(get_num_punctuations, dataset["reviewText"])))
punctuation_counts_df.columns = [f"punc_{col}" for col in punctuation_counts_df.columns]
additional_features = pd.concat([additional_features,punctuation_counts_df], axis = 1)
additional_features["ratio_uppercase_words"] = additional_features["num_uppercase_words"]/additional_features["num_words"]
additional_features["reviewId"] = dataset["reviewId"] 
```
